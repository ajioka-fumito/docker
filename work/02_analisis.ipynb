# XRD解析

## 1.1 データセットの準備

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import interpolate
from PIL import Image
%matplotlib inline
# init dirs
for i in glob.glob("./out/02_analisis/*.csv"):
    os.remove(i)
for i in glob.glob("./out/02_analisis/graph/*.png"):
    os.remove(i)
```

```python
prop = pd.read_csv("./out/01_fit/prop.csv",index_col=0)
pred = pd.read_csv("./out/01_fit/pred.csv")
orientations = [str(i) for i in list(prop.index)]
```

## <font color="crimson">1.2 **解析に使わない結晶方位を削除!!!!**</font>
remove_orientationのリストに直接書き込む  
ここでprop,predに関しても削除を行う.

```python
remove_orientation = ["111"]
for now in remove_orientation:
    orientations.remove(now)
    # propについて
    prop = prop.drop(int(now))
    # predについて
    pred = pred.drop([now+"theta",now+"intensity"],axis=1)
    
x0 = prop["x0"]
FWHM = prop["FWHM"]
```

## 1.3 **k&alpha;1の波長，C<sub>h00</sub>の定義**
測定方法，方位に依存した定数になるため異なるデータを解析する際は注意すること

```python
lam = 0.070931
C_h00 = 0.316
```

## 2. 各種パラメータの算出
### 2.1 Kの算出
$$
    K = \frac{2\sin(\theta)}{\lambda}
$$
に基づく．   
以下ではすべての方位をベクトルとして取扱うためKもベクトルとなるため注意

```python
theta = np.array(x0)/2
theta = np.pi*theta/180 # deg to rad
K = 2*np.sin(theta)/lam
```

### 2.2 &Delta;Kの算出
$$
    \Delta K = \frac{2\cos(\theta)\Delta\theta}{\lambda}
$$
に基づく

```python
delta_theta = np.array(FWHM)
delta_theta = np.pi*delta_theta/180 # deg to rad
delta_K = 2*np.cos(theta)*delta_theta/lam
K
delta_K
```

### 2.3 H<sup>2</sup>の算出
$$
    H^2 = \frac{(hk)^2+(kl)^2+(lh)^2}{(h^2+k^2+l^2)^2}
$$
に基づく

```python
H2 = []
for orientation in orientations:
    h,k,l = orientation
    h,k,l = int(h),int(k),int(l)
    h2 = ((h*k)**2+(h*l)**2+(k*l)**2)/(h**2+k**2+l**2)**2
    H2.append(h2)
H2 = np.array(H2)
```

### 2.4 &alpha;パラメータの決定

```python
coef = []
ls_alpha = []
for i in range(10**4):
    alpha = 2*i*10**(-7)
    y = (delta_K**2-alpha)/ K**2
    now = np.corrcoef(H2,y)[1][0]
    coef.append(now**2)
    ls_alpha.append(alpha)

coef = np.array(coef)
ls_alpha = np.array(ls_alpha)
idx = np.argmax(coef)
alpha = 2*idx*10**(-7)
plt.plot(ls_alpha*1000,coef,label="coefficient",color="crimson")
plt.legend()
plt.ylabel("coefficient")
plt.xlabel("alpha(10**4)")
plt.savefig("./out/02_analisis/graph/01_alpha.png")
plt.show()
```

### 2.5 &beta;A,&beta;Bの決定

```python
beta_B,beta_A = np.polyfit(H2,(delta_K**2-alpha)/K**2,1)
print("beta_A:",beta_A)
print("beta_B:",beta_B)
plt.scatter(H2,(delta_K**2-alpha)/K**2,color="mediumblue",label="data")
plt.plot(H2,beta_B*H2+beta_A,color="crimson",label="proximate")
plt.legend()
plt.ylim(min(0,min(beta_B*H2+beta_A)),max(beta_B*H2+beta_A)*2)
plt.ylabel("( delta_K^2 - alpha ) / K^2")
plt.xlabel("H^2")
plt.savefig("./out/02_analisis/graph/02_betaAB.png")
plt.show()
```

### 2.6 $q$値の決定

```python
q = -beta_B/beta_A
print("q:",q)
```

### 2.7 $\overline{C}$の決定

```python
C_bar = C_h00*(1-q*H2)
```

### 2.8 $A(L)$の算出

```python
A_L = pd.DataFrame()

for orientation in orientations:
    theta = np.array(pred[orientation+"theta"].dropna(how="all"))
    delta_theta = 2*(np.sin(theta[-1]*np.pi/360)-np.sin(theta[0]*np.pi/360))/lam
    delta_hlz = 1/delta_theta
    
    intensity = pred[orientation+"intensity"].dropna(how="all")
    intensity = intensity/max(intensity) # 正規化
    
    fft = abs(np.fft.fft(intensity))
    fft = fft/np.max(fft)
    hlz = np.arange(0,delta_hlz*len(fft)-0.001,delta_hlz)
    tck = interpolate.splrep(hlz,fft)
    fft_sp = interpolate.splev(np.arange(51),tck)
    A_L[orientation] = fft_sp
    plt.plot(np.arange(51),fft_sp)
plt.savefig("./out/02_analisis/graph/03_LvsAL.png")
plt.show()
```

### 2.8 $\ln{A(L)}$の算出

```python
lnA_L = A_L.apply(np.log)
```

### 2.9  $\ln{A^s(L)}$と$X(L)$の算出

```python
K2C = (K**2)*C_bar
lnAs_L,X_L = [],[]
for i in range(1,51):
    _,b,c = np.polyfit(K2C,lnA_L.loc[i],2)
    lnAs_L.append(c)
    X_L.append(b)
    plt.scatter(K2C,lnA_L.loc[i])
plt.xlabel("K^2*C_bar")
plt.ylabel("lnA(L)")
plt.savefig("./out/02_analisis/graph/04_K2CvslnA_L.png")
plt.show()
lnAs_L,X_L = np.array(lnAs_L),np.array(X_L)
```

### 3.10  $A^s(L)$の算出

```python
As_L = np.exp(lnAs_L)
```

### 3.11 転位密度$D$の算出

```python
a,b = np.polyfit(np.arange(1,5),As_L[0:4],1)
D = -b/a
print("D:",D)
plt.plot(np.arange(1,51),a*np.arange(1,51)+b,color="crimson",label="proximate")
plt.scatter(np.arange(1,51),As_L,color="mediumblue",label="Data")
plt.legend()
plt.ylim(0,1)
plt.ylabel("As(L)")
plt.xlabel("L")
plt.savefig("./out/02_analisis/graph/05_LvsAs_L.png")
plt.show()
```

### 3.11 $Re$の算出
$\ln(L)$と$X(L)/L^2$の近似直線より$Re$を算出する. 

```python
a,b = np.polyfit(np.log(np.arange(1,51))[20:],(X_L/np.arange(1,51)**2)[20:],1)
Re = np.exp(-b/a)
```

```python
fig = plt.figure()
fig.subplots_adjust(wspace=1)
ax1 = fig.add_subplot(121,xlabel="ln(L)",ylabel="X(L)/L^2")
ax1.set_title("Overall")
ax1.scatter(np.log(np.arange(1,51)),X_L/np.arange(1,51)**2,color="mediumblue",label="Data")
ax1.plot(np.linspace(0,4,100),a*np.linspace(3,4,100)+b,color="crimson",label="proximate")
ax1.set_ylim(min(X_L/np.arange(1,51)**2),max(X_L/np.arange(1,51)**2))
ax1.ticklabel_format(style='sci',axis='y',scilimits=(0,0))

ax2 = fig.add_subplot(122,xlabel="ln(L)",ylabel="X(L)/L^2")
ax2.set_title("Local")
ax2.scatter(np.log(np.arange(1,51))[20:],(X_L/np.arange(1,51)**2)[20:],color="mediumblue",label="Data")
ax2.plot(np.linspace(3,4,100),a*np.linspace(3,4,100)+b,color="crimson",label="proximate")
ax2.set_ylim(min((X_L/np.arange(1,51)**2)[20:]),max((X_L/np.arange(1,51)**2)[20:]))
ax2.ticklabel_format(style='sci',axis='y',scilimits=(0,0))
fig.savefig("./out/02_analisis/graph/06_lnLvsXL.png")
fig.show()
```
