# Main

## Fitting Function の定義
$$
    Lorentz function: f(x;x_0,\gamma) = \frac{1}{\pi}\frac{\gamma}{(x-x_{0})^2-\gamma^2}
$$
この関数は規格化されているので以下で定義する関数では倍率としてhを導入した．

```python
def Lorentz(x,params):
    """
    params = [h,x0,gamma]
    """
    y = 1/(np.pi*params[2]) * (params[2]**2/((x-params[1])**2 + params[2]**2))
    y = y/np.max(y)
    y = params[0]*y
    return y
```

上記関数の重ね合わせを行うための関数を定義

```python
def stack_funcs(x,*params):
    noize = params[0]
    ka1_params = params[1:(len(params)+1)//2]
    ka2_params = params[(len(params)+1)//2:]
    stack = Lorentz(x,ka1_params)+Lorentz(x,ka2_params)+noize
    return stack
```

## Sub Function の定義
初期値の決定に使用する関数群

```python
def bragg(*params):
    """
    params = [lamda_ka1,lamda_ka2,theta1,delta]
    """
    theta1 = params[2]*np.pi/180/2
    delta  = params[3]*np.pi/180
    return abs(params[1]/params[0]*np.sin(theta1)-np.sin(theta1+delta))
```

```python
def delta(ka,x0):
    ans = 10**10
    delta = 0
    for i in range(0,1000,2):
        now = i/1000
        sub = bragg(*[ka[0],ka[1],x0,now])
        if sub<ans:
            delta = now
            ans = sub
    return delta
```

```python
def noise(intensity):
    noise = sorted(intensity)
    noise = noise[:100]
    return np.mean(noise)
```

## **Fittingの実行**
### **1 初期値の決定**
フィッテイングを行うためには初期値を妥当なものにしておく必要がある．  
そのためヒストグラムから算出可能な初期値は算出を行い，その他の初期値に関してはよく採用されている値を採用した．  
ここで決定しておくべき初期値をまとめておくと以下の7つ{1, 3, 3}になる．  
{noize，ka1の初期値，ka2の初期値}
#### **1-1 初期 h_ka1,h_ka2 の決定**
最大intensityの0.8倍をh_ka1の初期値として採用する．  
-->大きな値を採用してしまうとka2が負の値をとってしまうため.  
h_ka2の初期値はh_ka1の0.5倍の値を採用している．
#### **1-2 x0の決定**
1-1における最大intensityに対応するthetaのidxからx0を決定する．
#### **1-3 &Delta;xの算出**
ka2のピークはka1のピークから少しずれることが知られている．  
実験原理より  
$$
    2d\sin(\theta) = n\lambda
$$
以下簡単のためn=1についてのみ考える．（それが一般的らしい）  
k&alpha;1,k&alpha;2についてそれぞれ上式を用いると
$$
    2d\sin(\theta_{k\alpha1}) = \lambda_{k\alpha1} \\
    2d\sin(\theta_{k\alpha2}) = \lambda_{k\alpha2}
$$
ここで&theta;<sub>k&alpha;2</sub> = &theta;<sub>k&alpha;1</sub> + &Delta;&theta; より
$$
    2d\sin(\theta_{k\alpha2}) = 2d\sin(\theta_{k\alpha1}+\Delta\theta) = \lambda_{k\alpha2}
$$
上式について&Delta;&theta;以外が既知であるため&Delta;&theta;は数値計算により近似値が計算可能である.  
ここで計算される&Delta;&theta;はブラック条件から算出されているためXRDの結果上では2&Delta;&theta;ずれるため2倍してから格納している
#### **1-4 Noiseの算出**
特に説明なし
#### **1-5 初期値リストの作成**

#### **1-6 Fittingの実行**


```python
plt.figure(figsize=(14,10))
ls_x0 = []
ls_FWHM = []
ls_predict = []
for i,orientation in enumerate(orientations):
    theta = data[orientation+"theta"].dropna(how="all")
    intensity = data[orientation+"intensity"].dropna(how="all")
    # 1-1, 1-2
    idx = np.argmax(np.array(intensity))
    x0,max_intensity = theta[idx],intensity[idx]
    # 1-3
    delta_x = delta(ka,x0)
    # 1-4
    noize = noise(intensity)
    # 1-5
    init_params = [noize, 0.8*max_intensity, x0        , 0.04,
                          0.4*max_intensity, x0+delta_x, 0.06]
    # 1-6
    popt,_ = curve_fit(stack_funcs,theta,intensity,p0 = init_params)
    plt.subplot(2,3,i+1)
    plt.plot(theta,intensity,color="mediumblue",label="Data")
    plt.plot(theta,stack_funcs(theta,*popt),color="crimson",label="Predict")
    plt.fill_between(theta,Lorentz(theta,popt[1:4])+popt[0],popt[0],color="green",label="ka1",alpha=1)
    plt.fill_between(theta,Lorentz(theta,popt[4:])+popt[0],popt[0],color="aqua",label="ka2",alpha=0.5)
    plt.legend()
    plt.title("orientation:"+orientation)
    plt.xlabel("theta")
    plt.ylabel("intensity")
    ls_x0.append(popt[2])
    ls_FWHM.append(popt[3]*2)
    ls_predict.append(Lorentz(theta,popt[1:4]))
plt.savefig("./out/01_fit/graph/02_predict.png")
plt.show()

# csvに出力
prop = pd.DataFrame(np.array([ls_x0,ls_FWHM]).T,columns=["x0","FWHM"],index=orientations)
prop.to_csv("./out/01_fit/prop.csv")

pred = pd.DataFrame()
for i,orientation in enumerate(orientations):
    theta = data[orientation+"theta"].dropna(how="all")
    pred = pd.concat([pred,theta,ls_predict[i]],axis=1)
columns = [ orientations[i//2]+"theta" if i%2==0 else orientations[i//2]+"intensity" for i in range(len(orientations)*2)]
pred.columns = columns
pred.to_csv("./out/01_fit/pred.csv",index=None)
```
